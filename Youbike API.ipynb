```python
#Accessing Taipei Youbike API
import requests
import json
import pandas as pd
import numpy as np

url= "https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json"

response= requests.get(url)
if response.status_code==200:
    data=json.loads(response.text)
else: print("error")
    
#Importing Data into DataFrame and Formatting Data Frame
df=pd.json_normalize(data)
df= df.rename(columns={'tot': 'Total Spaces', 
                       'sbi': 'Bikes available',
                       'bemp': 'Vacant Spaces',
                       'act': 'Station Status',
                       'snaen': 'Station'})
df= df.drop(['sno', 'sna','sarea', 'mday','ar','sareaen','aren', 'srcUpdateTime','updateTime','infoDate'], axis=1)
df= df.set_index('Station')

```

```python
filtered_data = df[df["Bikes available"]<4]
filtered_data
```

```python
#Creating Data frame with only Stations and Lat/Long Data
dflatlong=df.drop(['Total Spaces','Bikes available','Vacant Spaces','Station Status','infoTime'],axis=1)
dflatlong.to_csv('example.tsv', sep='\t')
```
